# Exercício Adaptado com a Base de Informações de Desempenho do Atleta Wesley Leite - QS/Qualifying Series - WSL

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_desempenho=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/1Desempenho QS.xlsx')
display(df_desempenho)
```

```python
fig=sns.displot(data=df_desempenho,x='Melhor Nota',kind='ecdf')
plt.show(fig)
```

```python
grafi_linha=sns.lineplot(df_desempenho,x='Ano',y='Posição',color='red')#lineplot comando monta gráfico de linhas
plt.show(grafi_linha)
```

```python
fig=sns.displot(data=df_desempenho,x='Ano',kind='kde')
plt.show(fig)
```

```python
plt.figure(figsize=(20,4)) #Gráfco Plot = Gráfico de Linha
plt.plot(df_desempenho['Ano'],df_desempenho['Melhor Nota'],label='Xapa',color='g',ls='--',lw='2')#Date é o eixo x e Close é o eixo y - comando label monta a legenda 'NYA' -
#color é o comando para definir a cor da linha - ls comando para tipo de linha e lw comando para espessura da linha
plt.legend(loc=2,fontsize=18) #plt.legend comando pra plotar a legenda. sem essa linha, a legenda não aparece
plt.ylabel('Melhor Nota')
plt.xlabel('Ano')
plt.title('Gráfico de Desempenho no Qualifying Series - Wesley Leite')
#plt.axis(xmin='2012',xmax='2024')
#plt.xticks(['2012','2014','2016','2018','2020','2022','2024'])
plt.xticks(rotation=90)
plt.show()#.show para visualizar gráfico
```
